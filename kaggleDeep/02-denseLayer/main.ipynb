```python
import matplotlib.pyplot as plt 
import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras import layers
import pandas as pd
```

```python
concrete =  pd.read_csv("../00-data/concrete.csv")
concrete.head()
print("the shape of concrete is {}".format(concrete.shape))
```

```python
# print(concrete.columns)
concrete.describe()
```

```python
model = keras.Sequential([
    layers.Dense(units = 512 , activation='relu',input_shape = (len(concrete.columns)-1,)),
    layers.Dense(units = 512 , activation='relu'),
    layers.Dense(units = 512 ,),
    layers.Dense(units = 1 , )
])
print(model.weights)
```

可选方案: 替代 RELU

“ relu”激活的变体有一个完整的家族——“ elu”、“ selu”和“ swish”等等——所有这些都可以在 Keras 使用。有时一个激活在给定的任务中会比另一个执行得更好，因此您可以在开发模型时考虑尝试激活。ReLU 激活在大多数问题上表现良好，所以这是一个很好的开始。


```python
activation_layer = layers.Activation('elu')

x = tf.linspace( -3.0 ,3.0 , 100 )
y = activation_layer(x)
```

```python
plt.figure(dpi=100)
plt.plot(x,y)
plt.xlim(-2,2)
plt.xlabel("input")
plt.ylabel("output")
plt.show()
```

```python
def drawOut(x,y):
    plt.figure(dpi=100)
    plt.plot(x,y)
    plt.xlabel("input")
    plt.ylabel("ouput")
    plt.show()
```
